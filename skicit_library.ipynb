# **simple examples using the Scikit-learn library**

#  Classification on Iris Dataset

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:, :2]
y = iris.target
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
pred = model.predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=pred)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Logistic Regression - Iris Dataset")

plt.show()
```

# **Linear Regression using Diabetes Dataset**

```python

import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
```

```python
diabetes = datasets.load_diabetes()
```

```python
X = diabetes.data[:, 2].reshape(-1,1)
y = diabetes.target
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
pred = model.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,pred)
plt.xlabel("Feature value")
plt.ylabel("Disease progression")
plt.title("Linear Regression using Diabetes Dataset")
plt.show()
```

**K-Means Clustering using Wine Dataset**

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans

wine = datasets.load_wine()

X = wine.data[:, :2]

kmeans = KMeans(n_clusters=3)

# TRAIN MODEL FIRST
kmeans.fit(X)

# THEN ACCESS THESE
labels = kmeans.labels_
centers = kmeans.cluster_centers_

plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centers[:,0], centers[:,1], marker='x', s=200)

plt.show()
```
